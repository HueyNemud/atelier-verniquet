```python
import pandas as pd


a = pd.read_excel(f"{LEFT}.ods", engine="odf", header=(0,1))
b = pd.read_excel(f"{RIGHT}.ods", engine="odf", header=(0,1))
```

```python

from IPython.display import display_html 

v = pd.DataFrame(a != b).applymap(lambda x: "background:red" if x != 0 else "")

to_side = lambda tbl : tbl.style.apply(lambda s: v, axis=None).set_table_attributes("style='display:inline'")

a =  to_side(a).set_caption(LEFT.upper())
b = to_side(b).set_caption(RIGHT.upper())

display_html(a._repr_html_()+b._repr_html_(), raw=True)
```
