```python
import pandas as pd 
import os

# Tous les angles aux sommets
sommets = pd.read_csv("toutes_les_bases.csv",header=[0,1], index_col=[0,1])

# Somme des angles des triangles, relevées dans le mémoire
total = pd.read_csv("angles_totaux.csv",header=[0,1], index_col=0)
```

```python
def dms2dec(d=0, m=0, s=0):
    return d + m/60 + s/3600

def dec2dms(dec=0):
    deg = int(dec)
    mins = (dec - deg)*60
    secs = (mins - int(mins))*60
    return deg, int(mins), secs


columns = sommets.columns.levels[0]
cols2dec = lambda row: {columns[0]:dms2dec(*row[0:3]), columns[1]:dms2dec(*row[3:6])}

# Angles aux sommets en degrés décimaux
sommets_dec = sommets.apply(cols2dec, axis=1,  result_type='expand')

                        # Sommes des angles en degrés décimaux
total_dec = total.apply(cols2dec, axis=1,  result_type='expand')
```

```python
actual = sommets_dec.reset_index()
actual = actual.groupby(actual.index // 3).sum()

# Sommes des angles aux sommets
actual.apply(lambda x: dec2dms(x[0]) + dec2dms(x[1]), axis=1,  result_type='expand')
```

```python
import numpy as np

expected = total_dec.reset_index()
diff = actual - expected
diff[np.abs(diff) < 10**-13] = 0 # Force les très petites valeurs à 0 
test = (diff == 0).all()
assert test["angles augmentés"] == True and test["angles observés"] == True
```
